# Classification For Machine Learning

Classification is a process that involves categorizing or assigning predefined labels to input data based on its features. The goal of classification is to learn a mapping function from the input features to the output labels, enabling the model to predict the label of new, unseen data points accurately.

In binary classification, the labels are often represented as y ∈ {0, 1}, where 0 represents the negative class and 1 represents the positive class. The function f(x) takes the input features x and predicts the corresponding label y. The mapping function can be denoted as P(y|x), which represents the probability of the label y given the input features x. The mapping function P(y|x) can be estimated using a probabilistic model such as logistic regression. Logistic regression models the log-odds (also known as the logit) of the probability using a linear combination of the input features, followed by a sigmoid function to convert it to a probability.

The logistic regression model can be represented as:

logit(P(y=1|x)) = β₀ + β₁x₁ + β₂x₂ + ... + βₚxₚ,

where β₀, β₁, β₂, ..., βₚ are the coefficients associated with each input feature x₁, x₂, ..., xₚ.

The logit function is defined as:

logit(p) = log(p / (1 - p)),

where p represents the probability.

In logistic regression, we can calculate the probability of the positive class (y=1) as:

P(y=1|x) = σ(β₀ + β₁x₁ + β₂x₂ + ... + βₚxₚ).

Similarly, the probability of the negative class (y=0) can be calculated as:

P(y=0|x) = 1 - P(y=1|x).

The classification task aims to find the optimal decision boundary that separates the input space into different regions corresponding to different classes. This decision boundary can be linear or nonlinear.

In multi-class classification, the task involves assigning predefined labels to input data where there are more than two possible classes. The goal is still to learn a mapping function that accurately predicts the label for each data point based on its features, but the formulation and approach differ from binary classification. To achieve this, we use the softmax function, which takes a set of scores or logits for each class and converts them into probabilities. These probabilities represent the likelihood of each class being the correct label. The class with the highest probability is then chosen as the predicted label for a given data point.

The softmax function can be defined as follows:

P(y=c|x) = exp(z_c) / (sum(exp(z_i)) for i=1 to C),

P(y=c|x) represents the probability of the class c given the input features x. The logits or scores for each class are denoted as z_c. The softmax function calculates the exponential of each logit and normalizes them by dividing by the sum of the exponentials of all the logits.

There are various machine learning algorithms commonly used for classification tasks. Logistic regression is a popular algorithm that models the probability of an input belonging to a specific class using the logistic function. It provides a straightforward approach to binary classification by thresholding the predicted probability. Decision trees are another widely used algorithm that recursively splits the input space based on feature values to create a tree-like structure. Each leaf node corresponds to a class label, allowing for intuitive interpretation and handling of both categorical and numerical features.

Random forests, an ensemble method, combine multiple decision trees and make predictions by aggregating the results of individual trees. This approach improves accuracy and robustness, making it particularly useful in scenarios with high-dimensional data or potential overfitting. Support Vector Machines (SVM) are effective for both linear and nonlinear classification problems. SVM aims to find the optimal hyperplane that separates the data points with the largest margin, enabling good generalization to unseen data. The kernel trick allows SVM to handle nonlinear decision boundaries by mapping the data into a higher-dimensional feature space.

Neural networks, especially deep learning models, have gained significant attention and achieved impressive results in classification tasks. These models consist of multiple layers of interconnected neurons, with each neuron performing a weighted computation. Neural networks learn the appropriate weights during the training process to capture complex patterns and relationships in the data. Deep learning architectures like convolutional neural networks (CNNs) have shown remarkable performance in image classification, while recurrent neural networks (RNNs) excel in sequential data, such as natural language processing tasks.

When imputing missing values, there are various strategies you can consider based on the nature of the data and the specific context.

Mean/Median Imputation: Replace missing values with the mean or median of the respective column. This method is suitable for numeric features and assumes that the missing values are missing at random (MAR).

Mode Imputation: For categorical features, you can impute missing values with the mode (most frequent value) of the column. This method is also appropriate when the missing values are MAR.

Based on the Dataset, We know that '__pha__', '__condition_code__', '__neo__' and '__class__' are object types. We assume then that these represent categorical features. Imputation will be done on these features using Mode.<br>
For the remainder of the dataset, Imputation will be done using the Mean method.

```python
# Impute 'object' type columns with mode
object_columns = ['condition_code', 'neo', 'pha', 'class']
df[object_columns] = df[object_columns].fillna(df[object_columns].mode().iloc[0])

# Impute remaining columns with mean
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())
```

### Normalization

Normalizing the dataset is not always necessary, but it can have benefits depending on the specific algorithm and the characteristics of the data. One key consideration is the scale of features. If the features in the dataset have different scales, normalization can help bring them to a similar scale. This can be important for algorithms that are sensitive to feature scale, such as gradient-based optimization algorithms or distance-based algorithms. Normalization can improve convergence speed and performance for these algorithms, as well as make the interpretation of model coefficients or feature importance more meaningful.

Both StandardScaler and MinMaxScaler can be used for binary classification.

StandardScaler: StandardScaler standardizes features by subtracting the mean and dividing by the standard deviation. It transforms the data to have zero mean and unit variance. This scaler is useful when the distribution of the features is approximately Gaussian (or normal) and there are no specific requirements on the range of the features. It is commonly used with algorithms that assume the features are normally distributed, such as logistic regression, support vector machines (SVM), and neural networks.

MinMaxScaler: MinMaxScaler scales features to a specified range, usually between 0 and 1. It preserves the shape of the original distribution and is suitable when the distribution of the features is not necessarily Gaussian and when preserving the relative relationships between features is important. This scaler is commonly used with algorithms that are sensitive to the scale of the features, such as k-nearest neighbors (KNN) and algorithms that use distance-based calculations.

### Using GridSearchCV to perform Hyperparameter tuning of the Bayes Classifier

To address these limitations, various strategies can be explored, such as exploring other algorithms specifically designed for imbalanced datasets (esemble methods). These methods combine multiple base models to create a stronger and more robust predictive model.

One popular ensemble method for imbalanced datasets is called "Boosting." Boosting algorithms, such as AdaBoost (Adaptive Boosting) and XGBoost (Extreme Gradient Boosting), focus on iteratively improving the classification performance by giving more weight to misclassified instances or difficult-to-classify instances. This way, the algorithm pays more attention to the minority class and tries to learn from its mistakes, leading to better overall performance.

Another ensemble method is "Bagging" (Bootstrap Aggregating), which trains multiple base models on different subsets of the dataset using bootstrapping. The final prediction is then made by aggregating the predictions of all the base models. Bagging can help reduce the variance and improve the generalization of the model.

Additionally, there are specialized ensemble methods designed for imbalanced datasets, such as "Random Forests" and "EasyEnsemble." Random Forests combine the principles of bagging and feature randomness to build an ensemble of decision trees. EasyEnsemble creates multiple subsets of the majority class to balance the class distribution and trains a separate classifier on each subset. The final prediction is obtained by aggregating the predictions of all the individual classifiers.
